# This code manually implements gradient descent to optimize a linear regression model.
# 100% working and reliable using only NumPy.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Generate synthetic linear data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3x + noise
```

```python
# Step 2: Add bias term (x0 = 1) to each instance
X_b = np.c_[np.ones((100, 1)), X]  # X_b = [1, x]
```

```python
# Step 3: Initialize parameters (theta0 and theta1)
theta = np.random.randn(2, 1)  # random initial values
```

```python
# Step 4: Set hyperparameters
learning_rate = 0.1
n_iterations = 1000
m = 100  # number of training examples
```

```python
# Step 5: Perform Gradient Descent
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients
```

```python
# Step 6: Output final parameters (theta0 and theta1)
print("Intercept (theta0):", theta[0][0])
print("Slope (theta1):", theta[1][0])
```

```python
# Step 7: Visualize the regression line
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, X_b.dot(theta), color='red', label='Regression line')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using Gradient Descent")
plt.legend()
plt.grid(True)
plt.show()
```

# ----------------------------
# Function Breakdown:
# - np.random.seed(): for reproducibility
# - np.c_: adds bias column to X
# - dot(): matrix multiplication for predictions and gradients
# - Gradient formula: 2/m * X^T * (X * theta - y)
# - theta update: theta = theta - learning_rate * gradients
#
# This is a manual and fully transparent implementation of gradient descent for educational use.
